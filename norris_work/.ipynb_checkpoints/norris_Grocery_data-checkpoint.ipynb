```python
### Setup 

import pandas as pd
from sqlalchemy import create_engine
```

```python
#### Store CSV into a DF

csv_file = (r"C:\Users\rnorris\Documents\grocery.data\Groceries_dataset.csv")
grocery_data_df = pd.read_csv(csv_file)
grocery_data_df.head()
```

```python
### Bring in another CSV file and turn into a DF

day_sell_csv = (r"C:\Users\rnorris\Documents\GitHub\etl-project\Resources\Day_sell_24_12_18.csv")
day_sell_df = pd.read_csv(day_sell_csv)
day_sell_df.head()
```

```python
## Clean up df by adjusting colums to portray appropriate data

# split out date columm data to have data after ";" placed into "zn" column

date_break = day_sell_df["Date"].copy().str.split(";",n = 1, expand = True )
date_break = date_break.rename(columns = {0:"Date", 1:"placeholder"})

date_break.head()
```

```python
# Merge back dfs

combined_df = pd.merge(day_sell_df, date_break, left_index = True, right_index = True)

combined_df.head()
```
