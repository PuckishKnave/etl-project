```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float, Date
from sqlalchemy.orm import Session

```

```python

### Bring in another CSV file and turn into a DF

day_sell_csv = (r"C:\Users\rnorris\Documents\GitHub\etl-project\Resources\day_sell.csv")
day_sell_df = pd.read_csv(day_sell_csv)
day_sell_df['Date'] = pd.to_datetime(day_sell_df.Date)
day_sell_df.head()
```

```python
# read in first grocery csv 
grocery_csv = "./Resources/Groceries_dataset.csv"
grocery_df = pd.read_csv(grocery_csv)

# renamed Member column
grocery_df.rename(columns={'Member_number':'MemberID'},
                 inplace=True)
grocery_df['Date'] = pd.to_datetime(grocery_df.Date)
grocery_df.index.name = 'Index'
grocery_df.head()
```

```python
## Merge

combined_df = pd.concat([grocery_df, day_sell_df])

combined_df.sort_values(by='Date',inplace = True, ascending = False)
combined_df.head()
```

```python
### Explanation

# Since no mutual date within data sets, queries seeking to find data based on a join of the dataframes will
# return results tht don't require the join in the first place. Essentially, there are no two of the same dates in either
# csv files. 
    
```
